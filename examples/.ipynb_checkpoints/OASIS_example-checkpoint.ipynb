## IDM example with OASIS policy data

IDM is a hierarchical representation of data. In the current version Accounts, Policies, and Insured Objects are individual JSON compliant documents. These documents are extensible and, most importantly, can be stored into a MongoDB database, which supports complex queries and aggregations on these data structures. 

```python
import sys
import pprint
from pymongo import MongoClient
sys.path.append("../src")
from load_OASIS import load_ktest_example
```

### Load an example policy into the hierarchical format

Load a ktest example from the OASIS project https://github.com/OasisLMF/ktest and convert it into a hierarchical structure. To showcase IDM additional parameters are added randomly.

```python
account, policy, ins_objs = load_ktest_example(20)
```

```python
pprint.pprint(policy)
```

### Initialize a MongoDB instance and load all ktest examples into it

Initialize a MongoDB instance. This can either be done
- locally
- on a dedicated server/cluster
- in the cloud, e.g. via MongoDB Atlas (https://www.mongodb.com/cloud/atlas)

Please refer to https://www.mongodb.com/ for further information.

```python
# initialze a client to a running MongoDB instance and select IDM as the database (it is created lazily)
client = MongoClient('localhost')
db = client["IDM"]
```

```python
# insert all available (correctly formated) OASIS example above documents into the database:
for acc_id in range(38):
    account, policy, ins_objs = load_ktest_example(acc_id)
    if account != None:
        #print(acc_id)
        try:
            db.accounts.insert_one(account)
            db.policies.insert_one(policy)
            db.insured_objects.insert_many(ins_objs)
        except:
            continue
```

Now the documents can be analyzed using MongoDB's framework, e.g. via [MongoDB Compass](https://www.mongodb.com/products/compass)

```python
from IPython.display import Image
Image("../data/Compass.png")
```


Of course it is also possible to use the Mongo's rich Python API to analyze and query the data.

```python
for hv_locs in db.insured_objects.find({"tiv":{"$gt": 100000000}}, {"acc_id":1, "tiv":1, "location.geometry":1}).sort("tiv",-1):
    pprint.pprint(hv_locs)
```
